```python
import pandas as pd
```

```python
df = pd.read_csv("coordinates", sep="\s+")
df['X'] = df['X'] * 2.54 / 100
df['Y'] = df['Y'] * 2.54 / 100
df['Z'] = df['Z'] * 2.54 / 100
```

```python
for row in df.itertuples():
    # print(row)
    print(f"tagsMap.put({row.ID}, new Translation3d({row.X:.3f}, {row.Y:.3f}, {row.Z:.3f}));")
```
